### Group By

```python
# Count
dfpMorosidad.groupby('tipo_vivienda').agg({'ID':'count'})
```

```python
# Count
dfpMorosidad.groupby('tipo_vivienda', as_index = False).agg({'ID':'count'})
```

```python
# Sum
dfpMorosidad.groupby('tipo_vivienda', as_index = False).agg({'ingreso':'sum'})
```

```python
# Avg
dfpMorosidad.groupby('tipo_vivienda', as_index = False).agg({'ingreso':'mean'})
```

```python
# Mas de una agrupacion
dfpMorosidad.groupby(['tipo_vivienda', 'nivel'], as_index = False).agg({'ingreso':'mean'})
```

### Case When

```python
def tramoEdad(edad):
    if edad >= 50:
        return 'Mayor'
    elif edad > 25:
        return 'Adulto'
    else:
        return 'Joven'
dfpMorosidad['edad'].apply(tramoEdad)
```

```python
# Revision
dfpMorosidad['edad'].apply(tramoEdad).value_counts()
```

```python
# SELECT A.*, B.*
# FROM CURSO A
# INNER JOIN INSTRUCTOR B
# ON A.idInstructor = B.idInstructor
# Forma 1
curso.merge(instructor, on=["idInstructor"])
```

```python
# SELECT A.nombre as NOMBRE_CURSO, B.NOMBRE as NOMBRE_PROFESOR
# FROM curso A
# inner join instructor B
# on A.idInstructor = B.idInstructor
curso[['idInstructor', 'nombre']].merge(instructor[['idInstructor', 'nombre']], on=["idInstructor"])\
    .drop(columns = 'idInstructor').rename(columns={'nombre_x':'NOMBRE_CURSO', 'nombre_y':'NOMBRE_PROFESOR'})
```

## 4.2.- Left Join

    Cuando las claves son iguales:
    df1.merge(df2, on=["clave"], how='left')
    Cuando las claves no son iguales:
    df1.merge(df2, left_on = "clave1", rigth_on = "clave2", how = "left")

```python
# SELECT A.*, B.*
# FROM CURSO A
# LEFT JOIN INSTRUCTOR B
# ON A.idInstructor = B.idInstructor
# Forma 1
curso.merge(instructor, on=["idInstructor"], how = 'left')
```

```python
# Cual es la diferencia?
curso.merge(instructor, on=["idInstructor"], how = 'inner')
```

## 4.3.- Rigth Join

    Cuando las claves son iguales:
    df1.merge(df2, on=["clave"], how='right')
    Cuando las claves no son iguales:
    df1.merge(df2, left_on = "clave1", rigth_on = "clave2", how = "right")

```python
# SELECT A.*, B.*
# FROM CURSO A
# LEFT JOIN INSTRUCTOR B
# ON A.idInstructor = B.idInstructor
# Forma 1
curso.merge(instructor, on=["idInstructor"], how = 'right')
```

```python
# Cual es la diferencia?
curso.merge(instructor, on=["idInstructor"], how = 'left')
```

## 4.4.- Full Join

    Cuando las claves son iguales:
    df1.merge(df2, on=["clave"], how='full_outer')
    Cuando las claves no son iguales:
    df1.merge(df2, left_on = "clave1", rigth_on = "clave2", how = "outer")

```python
# SELECT A.*, B.*
# FROM CURSO A
# LEFT JOIN INSTRUCTOR B
# ON A.idInstructor = B.idInstructor
# Forma 1
curso.merge(instructor, on=["idInstructor"], how = 'outer')
```
