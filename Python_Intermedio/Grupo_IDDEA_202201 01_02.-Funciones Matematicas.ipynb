## 1.- Funciones Básicas

### 1.1.- abs

Devuelve el valor absoluto de un número (entero o de coma flotante).

```python
print("El valor absoluto de 3 es:", abs(3))
print("El valor absoluto de -3 es:", abs(-3))
print("El valor absoluto de -2.5 es:", abs(-2.5))
```

### 1.2.- pow

Recibe dos argumentos, eleva el primero argumento a la potencia del segundo argumento.

```python
print(pow(3, 2))
print(pow(-4, 3))
print(pow(-7, 4))
```

### 1.3.- round

Redondea un número flotante a una precisión dada en dígitos decimal (por defecto 0 dígitos)

```python
print(round(3.345345))
print(round(34.34547, 3))
print(round(-7.345365, 4))
```

### 1.4.- bin

Devuelve una representación binaria de un número entero o entero long, es decir, lo convierte de entero a binario.

```python
print(bin(3))
print(bin(0))
print(bin(8))
```

### 1.5.- divmod

Debe recibir dos argumentos numéricos, y devuelve dos valores: resultado de la división entera, y el resto.

```python
print(divmod(5, 2))
print(f"El resultado de dividir 5/2 es: Cociente --> {divmod(5, 2)[0]}, Residuo --> {divmod(5, 2)[1]}")
print(divmod(-4, 3))
```

### 1.8.- max

Si recibe más de un argumento, devuelve el mayor de ellos.

```python
listValores = [12, 15, 10, 4, 29]
print(f"El maximo valor es {max(listValores)}.")
listValores = [0, -50, -10]
print("Maximo Valor:", max(listValores))
listValores = [-25, -10]
print("Maximo Valor:", max(listValores))
```

```python
## Probar la funcion con fracciones
```

### 1.9.- min

Tiene un comportamiento similar a max(), pero devuelve el mínimo.

```python
listValores = [12, 15, 10, 4, 29]
print(f"El minimo valor es {min(listValores)}.")
listValores = [0, -50, -10]
print("minimo Valor:", min(listValores))
listValores = [-25, -10]
print("minimo Valor:", min(listValores))
```

```python
## Probar la funcion con fracciones
```

### 1.10.- sum

## 2.- Funciones Matemáticas de la Librería "math"

```python
# Importamos la librería math
import math
```

### 2.1.- ceil

Retorna el valor superior de x, el número entero más pequeño que es mayor o igual que x.

```python
print(math.ceil(3.56))
print(math.ceil(3.06))
print(math.ceil(-3.98))
print(math.ceil(-3.06))
print(math.ceil(2))
```

### 2.2.- floor

Retorna el primer número entero menor o igual que x.

```python
print(math.floor(3.56))
print(math.floor(3.06))
print(math.floor(-3.98))
print(math.floor(-3.06))
print(math.floor(0))
```

### 2.3.- factorial

Retorna el factorial de x como un número entero.

```python
varNum = 5
print("El Factorial de", varNum, "es", math.factorial(varNum))
varNum = 8
print(f"El Factorial de {varNum} es {math.factorial(varNum)}.")
```

```python
## Probar el factorial negativo
```

### 2.4.- exp

Retorna e elevado a la x potencia, dónde e = 2.718281… es la base de los logaritmos naturales.

```python
print(math.exp(0))
print(math.exp(1))
print(math.exp(-5))
print(math.exp(0.5))
print(math.exp(-0.5))
```

### 2.5.- log

Con un argumento, retorna el logaritmo natural de x (en base e).

```python
print(math.log(1))
print(math.log(2.718281828459045))
print(math.log(0.006737946999085467))
print(math.log(1.6487212707001282))
print(math.log(0.6065306597126334))
print(math.log(9, 3))
print(math.log(1/4, 2))
```

```python
## Probar logaritmos negativos
```

### 2.6.- sqrt

Retorna la raíz cuadrada de x.

```python
print(math.sqrt(0))
print(math.sqrt(1))
print(math.sqrt(1/4))
print(math.sqrt(1.6487212707001282))
print(math.sqrt(100))
```

### 3.5.- shuffle

Baraja el orden de las listas, matrices, tuplas.

```python
varListaNumeros = [1, 3, 5, 7, 8]
print(varListaNumeros)
print("Barajando los numeros")
random.shuffle(varListaNumeros)
print(varListaNumeros)
print("\n")
varListaCadenas = ['Aldo', 'Beto', 'Carlos', 'Daniel', 'Edgar']
print(varListaCadenas)
print("Barajando los nombres")
random.shuffle(varListaCadenas)
print(varListaCadenas)
```

### 3.6.- sample

Obtiene aleatoriamente n elementos de listas, matrices, tuplas.

```python
print(random.sample(["Zhang San","Yang Guo","Pequeña niña dragón","Guo Jing"], 1))
print(random.sample(["Zhang San","Yang Guo","Pequeña niña dragón","Guo Jing"], 2))
print(random.sample(["Zhang San","Yang Guo","Pequeña niña dragón","Guo Jing"], 3))
```
