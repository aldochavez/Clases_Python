# Analisis Exploratorio de Datos

```python
# Importamos la libreria pandas para el manejo de data estructurada
import pandas as pd
# Importamos la libreria Numpy para el manejo de matrices
import numpy as np
```

```python
# Importacion de los Datos
df = pd.read_csv('https://raw.githubusercontent.com/aldochavez/datasets/main/AdquisicionCreditoHipotecario.csv')
df.columns = ['ID', 'Genero', 'Estado_Civil', 'Dependientes', 'Educacion',
       'Independiente', 'Sueldo_Dependiente', 'Sueldo_Independiente', 'Monto_Prestamo',
       'Tiempo_Prestamo', 'Historial_Crediticio', 'Zona', 'Estado_Prestamo']
df.head()
```

**Resumimos el analisis con una funcion iterativa**

```python
#Analizando las variables categoricas
for variable in ['Genero','Estado_Civil','Independiente','Dependientes','Educacion','Zona','Estado_Prestamo']:
  print("#"*20, "Analizando la variable", variable, "#"*20 )
  display(df[variable].value_counts())
  display(df[variable].value_counts(normalize = True))
  print("\n")
```

**Resumimos el analisis con una estructura iterativa y el uso de funciones**

```python
#Definimos la funcion
def analisisConcentracion(variable):
  print("#"*20, "Analizando la variable", variable, "#"*20 )
  display(df[variable].value_counts())
  display(df[variable].value_counts(normalize = True))
  print("\n")
```

```python
#Analizando las variables categoricas
for variable in ['Genero','Estado_Civil','Independiente','Dependientes','Educacion','Zona','Estado_Prestamo']:
  analisisConcentracion(variable)
```

### Uniendo Grficos

```python
#Analizando la variable Genero
df['Genero'].value_counts().plot.bar()
plt.ylim(0,500)
plt.twinx()
df['Genero'].value_counts(normalize = True).plot(color='red')
plt.ylim(0,1)
#plt.savefig('Imagen1.jpg')
```

```python
#Analizando la variable Genero
df['Genero'].value_counts().plot.bar()
plt.ylim(0,500)
for i, text in enumerate(df['Genero'].value_counts()):
    plt.annotate(text, (i, text))
plt.twinx()
df['Genero'].value_counts(normalize = True).plot(color='red')
for i, text in enumerate(df['Genero'].value_counts(normalize = True)):
    plt.annotate(text, (i, text))
plt.ylim(0,1)
#plt.savefig('Imagen1.jpg')
```

```python
# Usarlo para las variables cateogricas mediante una estructura iterativa
for variable in ['Genero','Estado_Civil','Independiente','Dependientes','Educacion','Zona','Estado_Prestamo']:
  print("Graficando la variable:", variable)
  df[variable].value_counts().plot.bar()
  plt.twinx()
  df[variable].value_counts(normalize = True).plot(color='red')
  plt.ylim(0,1)
  plt.show()
```
