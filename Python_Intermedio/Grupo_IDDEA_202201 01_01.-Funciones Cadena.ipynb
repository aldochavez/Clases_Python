# Funciones Cadena

Los **string** o cadenas poseen diversas funciones que nos van a permitir trabajar de una mejor maneras con estos tipos de variables

## 1.- Funciones Básicas

### 1.1.- upper

Convierte todo el string a mayúsculas

```python
varMinuscula = 'uefa'
print('La variables en mayúscula es:', varMinuscula)
print("Conviertiendo a mayúscula...")
print('La variables en mayúscula es:', varMinuscula.upper())
```

```python
## Definir una variable con su nombre en minuscula y mostrarla en mayuscula usando la funcion upper:
```

### 1.2.- lower

Convierte todo el string a minúsculas

```python
varMayuscula = 'oLa PrInCesItaH, cOmO EsTaS'
print('La variable en minuscula es:', varMayuscula)
print("Conviertiendo a Minuscula...")
print('La variable en minuscula es:', varMayuscula.lower())
```

```python
## Definir una variable con su apellido en mayuscula y mostrarla en minuscula usando la funcion lower:
```

### 1.3.- capitalize

Coloca la primera letra en mayúscula y el resto en minúscula

```python
varString = 'rayuela'
print('La variable es:', varString)
print("Usando la funcion Capitalize...")
print('La variable en utilizando la funcion capitalize es:', varString.capitalize())
```

```python
## Definir una variable con el nombre de su libro preferido y mostrarlo utilizando la funcion capitalize:
```

### 1.4.- title

### 1.6.- center

Centra la cadena con la longitud proporcionada y el carácter. Si no se especifica ningún carácter, el valor predeterminado es ' ' (espacio).

```python
varCenter = 'Aldo'
print(varCenter.center(16))
varCenter = 'Pedro'
print(varCenter.center(16, '.'))
varCenter = 'Josefina'
print(varCenter.center(16, '-'))
varCenter = 'Alessandra'
print(varCenter.center(16, '_'))
```

### 1.9.- strip

Elimina cualquier carácter especificado desde el principio y el final de la cadena. Si no se pasa ningún argumento, el valor predeterminado es ' ' (espacio).

```python
varString = '     Aldo       '
print(varString.strip())
varString = '______Jose_______'
print(varString.strip('_'))
varString = 'alalalalalalalalalPedroalalal'
print(varString.strip('al'))
```

```python
## Retirar los signos de puntuacion de la siguiente variable
varString = '.,palabra,.'
```

### 1.10.- lstrip

Funciona exactamente como strip() con la excepción de que elimina caracteres solo del lado izquierdo de la cadena.

```python
varString = '     Aldo       '
print(varString.lstrip())
varString = '______Jose_______'
print(varString.lstrip('_'))
varString = 'alalalalalalalalalPedroalalal'
print(varString.lstrip('al'))
```

```python
## Retirar los signos de puntuacion de la parte izquierda de la siguiente variable
varString = '.,palabra,.'
```

### 1.11.- rstrip

Funciona exactamente como strip() con la excepción de que elimina caracteres solo del lado derecho de la cadena.

```python
varString = '     Aldo       '
print(varString.rstrip())
varString = '______Jose_______'
print(varString.rstrip('_'))
varString = 'alalalalalalalalalPedroalalal'
print(varString.rstrip('al'))
```

```python
## Retirar los signos de puntuacion de la siguiente variable
varString = '.;palabra,.'
```

### 2.9.- find

Devuelve la posición inicial del carácter o cadena especificados. Si su argumento no está presente en la cadena, devolverá -1.

```python
sentence = 'This is a sentence find'
print("La palabra 'i' esta en la posicion:", sentence.find('i'))
print("La palabra 'zz' esta en la posicion:", sentence.find('zz'))
print("La palabra 'i' esta en la posicion:", sentence.find('i', 4))
print("La palabra 'i' esta en la posicion:", sentence.find('i', 6, 15))
```

## 4.- Tips Adicionales

### 4.1.- Multiplicación de Caracteres

Si se multiplica un caracter por un número se repetira tantas veces se indique

```python
varString = 'Hola'
print('*' * 3)
print(varString * 10)
print(' ' * 20 + 'Titulo' + ' ' * 20)
```

### 4.2.- Uso de Format para definir cadenas de caracteres

Se pueden incluir variables para construir cadenas

```python
print("Hola mi nombre es {}".format('Aldo'))
print("Hola mi nombre es {} y mi apeliido es {}.".format('Aldo', 'Chavez'))
varNombre = 'Jose'
print("Mucho gusto Sr. {}.".format(varNombre))
```

### 4.3.- Uso de F-String para definir cadenas de caracteres

Se pueden incluir variables para construir cadenas

```python
print(f"Hola mi nombre es {'Aldo'}")
print(f"Hola mi nombre es {'Aldo'} y mi apeliido es {'Chavez'}.")
varNombre = 'jose'
print(f"Mucho gusto Sr. {varNombre.capitalize()}.")
```

**Nota: para usar caracteres especiales usamos el caracter de escape '\'**

```python
print(f"Mucho gusto \"Sr.\" {varNombre.capitalize()}.")
```
