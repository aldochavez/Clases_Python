# Funciones Fechas

En Python, la fecha y la hora no son un tipo de datos en sí mismos, pero se puede importar un módulo llamado datetime para trabajar con la fecha y la hora. El módulo Python Datetime viene integrado en Python, por lo que no es necesario instalarlo externamente.
Las Clases Principales del Modulo Datetime son:

**date**: Una fecha que contiene los atributos año, mes y día.

**tiempo**: Una variable de tiempo cuyos atributos son hora, minuto, segundo, microsegundo.

**datetime**: Una combinación de date y time con los atributos año, mes, día, hora, minuto, segundo, microsegundo.

**timedelta**: Expresa la diferencia entre dos instancias de fecha, hora o fecha y hora con una resolución de microsegundos.

## 1.- Clase Date

**La clase Date maneja la siguiente --> sintaxis datetime.date(año, mes, día)**

```python
# Importamos la clase date
from datetime import date
```

### 1.1.- Construyendo nuestra fecha

La clase de date se usa para instanciar objetos de fecha en Python. Cuando se instancia un objeto de esta clase, representa una fecha en el formato AAAA-MM-DD. El constructor de esta clase necesita tres argumentos obligatorios año, mes y dia.

```python
varDate = date(2022, 1, 23)
print("La fecha actual es:", varDate)
```

```python
## Mostar la fecha de su cumpleaños
```

### 1.2.- today

Devuelve la fecha actual

```python
varDate = date.today()
print("La fecha actual es:", varDate)
```

### 1.3.- Atributos year, month y day

Nos permite obtener los atributos de año, mes y dia del objeto de fecha.

```python
varDate = date.today()
print("La fecha actual es:", varDate)
print("La año de la fecha actual es:", varDate.year)
print("El mes de la fecha actual es:", varDate.month)
print("El día de la fecha actual es:", varDate.day)
```

```python
## Obtener el año, mes y día de la fecha de tu cumnpleaños
```

### 1.4.- isoformat

Convierte un objeto de fecha en una cadena de caractres

```python
varFecha = date.today()
print("Variable Fecha:", varFecha)
print("Convirtiendo a string...")
print("Variable Fecha:", date.isoformat(varFecha))
print(date.isoformat(varFecha).isalnum())
```

```python
## Convierta la fecha de hoy (today) al formato DDMMYY usando funciones de cadena.
```

### 1.5.- fromtimestamp

Podemos crear objetos de fecha a partir de marcas de tiempo, la marca de tiempo es el número de segundos desde el 1 de enero de 1970 en UTC hasta una fecha determinada.

```python
varDate = date.fromtimestamp(1887639468)
print("La fecha es:", varDate)
varDate = date.fromtimestamp(0)
print("La fecha es:", varDate)
varDate = date.fromtimestamp(60 * 60 * 24)
print("La fecha es:", varDate)
varDate = date.fromtimestamp(60 * 60 * 24 * 31 * 12)
print("La fecha es:", varDate)
```

## 2.- Clase Time

**La clase Time maneja la siguiente -->  sintaxis es datetime.time(hora=0, minuto=0, segundo=0, microsegundo=0)**

```python
# Importamos la clase time
from datetime import time
```

### 2.1.- Construyendo nuestro time

La clase de time crea el objeto de tiempo que representa la hora local, independientemente de cualquier día

```python
varTime = time(13, 24, 56)
print("La hora actual es:", varTime)
varTime = time(hour = 21)
print("La hora actual es:", varTime)
varTime = time(minute = 46)
print("La hora actual es:", varTime)
varTime = time(second = 15)
print("La hora actual es:", varTime)
varTime = time(microsecond = 123456)
print("La hora actual es:", varTime)
varTime = time()
print("La hora actual es:", varTime)
```

```python
## Mostar hora actual
```

### 2.2.- Atributos hour, minute, second y microsecond

Nos permite obtener los atributos de hora, minuto, segundo y microsegundo del objeto de time.

```python
varTime = time(13, 24, 56, 123456)
print("La fecha actual es:", varTime)
print("La hora es:", varTime.hour)
print("Los minutos son:", varTime.minute)
print("Los segundos son:", varTime.second)
print("Los microsegundos son:", varTime.microsecond)
```

```python
# Mostrar las horas, minutos y segundos de la fecha actual
```

### 2.3.- isoformat

Convierte un objeto de fecha en una cadena de caractres

```python
varTime = time(13, 24, 56, 123456)
print("Variable de tiempo:", varTime)
print("Convirtiendo a string...")
print("Variable de tiempo:", time.isoformat(varTime))
print(time.isoformat(varTime).isalnum())
```

```python
## Convierta la hora de hoy (today) al formato hora-minutos-segundo usando funciones de cadena.
```

## 3.- Clase Datetime

**La clase datetime maneja la siguiente --> sintaxis es datetime.datetime(año, mes, día, hora=0, minuto=0, segundo=0, microsegundo=0)**

```python
# Importamos la clase datetime
from datetime import datetime
```

## 4.- Clase Timedelta

La clase timedelta de Python se usa para calcular las diferencias en las fechas y también se puede usar para la manipulación de fechas en Python. Es una de las formas más fáciles de realizar manipulaciones de fechas.

```python
# Importamos la clase timedelta
from datetime import timedelta
```

### 4.1.- Construyendo nuestro timedelta

**Sintaxis de la clase Timedelta --> datetime.timedelta(días=0, segundos=0, microsegundos=0, milisegundos=0, minutos=0, horas=0, semanas=0)**

```python
varTimedelta = timedelta(days = 2)
print("La diferencia de dias es:", varTimedelta)
varTimedelta = timedelta(hours = 21)
print("La diferencia de dias es:", varTimedelta)
varTimedelta = timedelta(minutes = 36)
print("La diferencia de dias es:", varTimedelta)
varTimedelta = timedelta(seconds = 45)
print("La diferencia de dias es:", varTimedelta)
varTimedelta = timedelta()
print("La diferencia de dias es:", varTimedelta)
varTimedelta = timedelta(days = 2, hours = 21, minutes = 36, seconds = 45)
print("La diferencia de dias es:", varTimedelta)
```

### 4.2.- Atributos day y second

Nos permite obtener los atributos de dias y segundo del objeto de timedelta.

```python
varTimedelta = timedelta(days = -2, hours = 21, minutes = 36, seconds = 45)
print("La diferencial es:", varTimedelta)
print("Los dias de diferencia son:", varTimedelta.days)
print("Los dias de diferencia son:", varTimedelta.seconds)
```

## 5.- Operaciones Frecuentes

### 5.1.- Resta de Fechas

```python
t1 = date(year = 2018, month = 7, day = 12)
t2 = date(year = 2017, month = 12, day = 23)
t3 = t1 - t2
print("t3 =", t3)
```

```python
# Veamos el tipo de dato que se obtiene al restar dos fechas
print(type(t3))
# Como vemos es un timedelta.
print("La cantidad de días de diferencia es:", t3)
print("La cantidad de días de diferencia es:", t3.days)
```

```python
# Obtener la cantidad de dias que pasaron desde hoy hasta su fecha de graduacion de universidad
```

### 5.2.- Sumar dias a una fecha

```python
varFechaActual = datetime.now()
print("Fecha Actual:", varFechaActual)
# La siguiente clase es dentro de siete dias
varFechaNueva = varFechaActual + timedelta(days = 7)
print("Fecha Proxima Clase:", varFechaNueva)
```

```python
# Obtener la fecha y hora de termino de la clase
varFechaActual = datetime(2022, 1, 23, 14)
print("Fecha Actual:", varFechaActual)
# La siguiente clase es dentro de siete dias
varFechaNueva = varFechaActual + timedelta(hours = 3)
print("Fecha Proxima Clase:", varFechaNueva)
```

**Nota: si sumanos un entero a una fecha nos dara error**
